```python
model = joblib.load("../artifacts/final_model.pkl")
encoder = joblib.load("../artifacts/target_encoder.pkl")

new_data = pd.DataFrame({
    "building_construction_year": [2026],
    "building_total_floors": [14],
    "apartment_floor": [12],
    "apartment_rooms": [2],
    "apartment_bedrooms": [3],
    "apartment_bathrooms": [2],
    "apartment_total_area": [160],
    "apartment_living_area": [140],
    "country": ["UAE"],
    "location": ["Dubai, UAE"]
})
new_data[["country" , "location"]] = encoder.transform(new_data[["country", "location"]])

y_pred_log = model.predict(new_data)[0]
y_pred_real = np.expm1(y_pred_log)

print("Predicted Log:", y_pred_log)
print("Predicted Price USD:", y_pred_real)
```
